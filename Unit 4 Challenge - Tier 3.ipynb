#### 1.2.  Loading the data

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
properties_t = properties.transpose()
properties_t = properties_t.reset_index()
new_header = properties_t.iloc[0] #grab the first row for the header
properties_t = properties_t[1:] #take the data less the header row
properties_t.columns = new_header #set the header row as the df header
properties_t.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# rename columns 'Unamed: 0' and 'NaN' as 'borough' and 'id', respectively
properties_t = properties_t.rename(columns={'Unnamed: 0': 'borough', pd.NaT: 'id'})
properties_t.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# melt propeties_t so all date columns are now in one row
properties_m = pd.melt(properties_t, id_vars= ['borough', 'id'])

# rename '0' column as 'month'  and 'value' as 'avg_price'
properties_m = properties_m.rename(columns = {0: 'month', 'value': 'avg_price'})
properties_m.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# convert avg_price to floar
properties_m['avg_price'] = pd.to_numeric(properties_m['avg_price'])
properties_m.dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
print(properties_m.count()) # get info on data count per column
print(properties_m['borough'].unique()) # get list of unique borough data rows
```

```python
properties_drop_na = properties_m.copy() # create new df to start removing na values
properties_drop_na = properties_drop_na.dropna() # drop all na values
print(properties_drop_na.shape)
print(properties_drop_na['borough'].unique())
```

```python
# create list of locations in 'borough' column that aren't actually boroughs. * I kept 'City of London' even though it's not an official borough
locs_to_remove = ['Inner London',
       'Outer London','NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER',
       'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND', 'LONDON',
       'SOUTH EAST', 'SOUTH WEST', 'England']
# create new df that keeps boroughs that are not (~) in the locs_to_remove list
properties_drop_loc = properties_drop_na[~properties_drop_na['borough'].isin(locs_to_remove)]
print(properties_drop_loc.shape)
print(properties_drop_loc['borough'].unique())
```

```python
# set final working dataframe name as 'df'
df = properties_drop_loc.copy()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
greenwich_prices = df[df['borough'] == 'Greenwich'] # subset df with only Greenwich data
greenwich_plot = greenwich_prices.plot(kind='line', x = 'month', y = 'avg_price') # create line plot
plt.title('Greenwich Average Housing Prices over Time')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.show()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
df_g = df.groupby(['borough', 'year']).mean() # creates new df (df_g) that consolidates data into average price per year for each borough
df_g.head(30)
```

```python
df_g = df_g.reset_index()
df_g.head()
```

```python
#create function to calculate price ratio
def price_ratio(p):
    d = df_g[df_g['borough'] == p]
    year2002 = float(d['avg_price'][d['year'] == 2002])
    year2022 = float(d['avg_price'][d['year'] == 2022])
    ratio = year2022 / year2002
    return ratio
```

```python
print(price_ratio('Barking & Dagenham')) # test price ratio function
```

```python
dict_ratios = {} # create empty list
for b in df_g['borough'].unique(): # iterate through each unique borough in df_g
    dict_ratios[b] = (price_ratio(b)) # add new entry to dict_ratios with the key = bourough and value = price ratio (calculated through price_ratio function)
print(dict_ratios)
```

```python
df_ratios = pd.DataFrame(dict_ratios, index=[0]) # convert dict_ratios into a dataframe
print(df_ratios.head())
```

```python
df_ratios_t = df_ratios.transpose()# transpose df_ratios so each ratio for each borough is on a single row
df_ratios_t.reset_index(inplace=True) # reset index
df_ratios_t = df_ratios_t.rename(columns={'index': 'Borough', 0: 'Price Ratio (2022/2002)'}) #rename columns
df_ratios_t.head()
```

```python
top10_df = df_ratios_t.sort_values('Price Ratio (2022/2002)', ascending=False).head(10) # create df for top 10 boroughs sorted by price ratios
top10_df.reset_index(drop=True, inplace=True)# reset index
print(top10_df)
top10_df['Price Ratio (2022/2002)'].mean()
```
